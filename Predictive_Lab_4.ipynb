```python
# Sample DataFrame with missing values
data = {'A': [1, 2, None, 4], 'B': [5, None, None, 8], 'C': [9, 10, 11, None]}
df = pd.DataFrame(data)

# Identify missing values using isnull() or isna()
missing_values = df.isnull()  # or df.isna()
print(missing_values)
```

```python
# Fill missing values with a specific value (e.g., 0)
df_filled = df.fillna(0)

# Drop rows with missing values
df_dropped = df.dropna()

# Interpolate missing values (useful for numerical data)
df_interpolated = df.interpolate()

print(df_filled)
print(df_dropped)
print(df_interpolated)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Sample numerical data
numerical_data = df[['A', 'B']].fillna(0)

# Min-Max Scaling
scaler = MinMaxScaler()
df_min_max_scaled = pd.DataFrame(scaler.fit_transform(numerical_data), columns=['A', 'B'])

# Z-Score Standardization
scaler = StandardScaler()
df_z_score_scaled = pd.DataFrame(scaler.fit_transform(numerical_data), columns=['A', 'B'])

print(df_min_max_scaled)
print(df_z_score_scaled)
```

```python
# Sample categorical data
df_cat = pd.DataFrame({'Category': ['A', 'B', 'A', 'C']})

# Create dummy variables
df_dummies = pd.get_dummies(df_cat, prefix='Category')

print(df_dummies)
```

Aggregation and Grouping


```python
# Sample DataFrame
data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 15, 10, 20, 10]}
df = pd.DataFrame(data)

# Group by 'Category' and calculate summary statistics
grouped = df.groupby('Category').agg(['mean', 'median', 'count'])

print(grouped)
```

```python
# Sample DataFrame
data = {'Category': ['A', 'A', 'B', 'B'],
        'SubCategory': ['X', 'Y', 'X', 'Y'],
        'Value': [10, 20, 30, 40]}
df = pd.DataFrame(data)

# Create pivot table
pivot_table = df.pivot_table(values='Value', index='Category', columns='SubCategory', aggfunc='mean')

print(pivot_table)
```

```python
# Sample DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'A': ['A0', 'A1', 'A3'], 'C': ['C0', 'C1', 'C3']})

# Concatenation
df_concat = pd.concat([df1, df2], axis=0)  # Vertical (default)
df_concat_hor = pd.concat([df1, df2], axis=1)  # Horizontal

# Merge (Inner join by default)
df_merged = pd.merge(df1, df2, on='A')

# Join (uses the index by default)
df_joined = df1.join(df2.set_index('A'), on='A')

print(df_concat)
print(df_merged)
print(df_joined)
```

```python
# Inner Join (default in merge)
df_inner = pd.merge(df1, df2, on='A', how='inner')

# Outer Join
df_outer = pd.merge(df1, df2, on='A', how='outer')

# Left Join
df_left = pd.merge(df1, df2, on='A', how='left')

# Right Join
df_right = pd.merge(df1, df2, on='A', how='right')

print(df_inner)
print(df_outer)
print(df_left)
print(df_right)
```
