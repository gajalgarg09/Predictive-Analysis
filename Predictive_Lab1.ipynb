```python
# Q6

fruits = ["apple" , "banana" , "grapes" , "watermelon"]

# accessing using index
print(fruits[0: len(fruits)])

#modifying
fruits[2] = "mango"
print(fruits)

# adding and removing
fruits.insert(4, "guava")
print(fruits)

fruits.remove("banana")
print(fruits)

 # length of list
print(len(fruits))

# sorting
fruits.sort()
print(fruits)

fruits.sort(reverse = True)
print(fruits)
```
